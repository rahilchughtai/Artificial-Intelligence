The notebooks in this directory should be read in the following order:
* [1-Tic-Tac-Toe](./1-Tic-Tac-Toe.ipynb) is an implementation of the game [tic-tac-toe](https://en.wikipedia.org/wiki/Tic-tac-toe).
* [2-Tic-Tac-Toe-Bitboard](./2-Tic-Tac-Toe-Bitboard.ipynb) is an implementation of tic-tac-toe that is based on [bitboards](https://en.wikipedia.org/wiki/Bitboard).
* [3-Minimax](./3-Minimax.ipynb) describes the *minimax algorithm*.
* [4-Memoization](./4-Memoization.ipynb) describes the programming techique of [memoization](https://en.wikipedia.org/wiki/Memoization).  This is a programming technique used to speed up computations that repeatedly have to evaluate a function for the same set of arguments.
* [5-Alpha-Beta-Pruning](./5-Alpha-Beta-Pruning.ipynb) implements [Alpha-Beta pruning](https://en.wikipedia.org/wiki/Alpha-beta_pruning).  *Memoization* is only added in a naive way.
* [6-Alpha-Beta-Pruning-Memoization](./6-Alpha-Beta-Pruning-Memoization.ipynb) implements *Alpha-Beta pruning* with a sophisticated version of *memoization*.
* [7-Game](./7-Game.ipynb) implements *Alpha-Beta pruning* with *memoization* and *progressive deepening*.
